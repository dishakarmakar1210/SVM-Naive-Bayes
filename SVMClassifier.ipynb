```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_classification
X,y =make_classification(n_samples=1000, n_features=2, n_redundant=0, random_state=42,n_classes=2, n_clusters_per_class=2)
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y, palette='Set1')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
```

```python
clf.fit(X_train, y_train)
```

```python
clf.coef_
```

```python
y_pred= clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.1, 0.2, 1, 2, 3, 10, 50, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': [1,0.1, 0.01, 0.2 ]
}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3, n_jobs=-1, cv=5).fit(X_train, y_train)
```

```python
grid
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.best_estimator_.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
